# Glider Problem Example #

## Problem Statement ##
We want to maximize the horizontal range of a glider.

### Variables and Inputs ###
$x(t)$ = Horizontal position of glider (Initially = 0 m)  
$v_x(t)$ = Horizontal component of $v_r$ (Initially = 13.23 m/s, Terminally = 13.23 m/s)  
y(t) = Vertical position of glider (Initially = 1000 m, Terminally = 900 m)  
$v_y(t)$ = Vertical component of $v_r$ (Initially = -1.288 m/s, Terminally = -1.288 m/s)  
$m$ = mass of glider with pilot (100 kg for this problem)  
$S$ = wing area of glider (14 m for this problem)  
$\rho$ = air density (1.13 $kg/m^3$)  
$g$ = acceleration due to gravity (9.81 $m/s^2$)  
$c_L(t)$ = lift coeficient of glider ($0 \le c_L \le c_{Lmax} = 1.4$)  
$T$ = time for glider to complete flight (unknown)  

There is also a thermal 250 meters ahead of our current position. The thermal provides an upward wind velocity that is modeled by the following function:
$$u_a(x) = u_me^{-(\frac{x}{R}-2.5)^2}(1-(\frac{x}{R}-2.5)^2)$$  
For this problem we set:  
$u_m = 2.5 m/s$,  
$c_0 = 0.034$,  
$k = 0.069662$.

### Problem Formulation: ###  
$$maximize \quad x(T)$$
Subject to:  

$$\frac{d}{dt}x(t) = v_x(t),\quad t \in [0,T]$$
$$\frac{d}{dt}v_x(t) = \frac{1}{m}(-Lsin(n(t)) - Dcos(n(t))),\quad t \in [0,T]$$
$$\frac{d}{dt}y(t) = v_y(t),\quad t \in [0,T]$$
$$\frac{d}{dt}v_y(t) = \frac{1}{m}(Lcos(n(t)) - Dsin(n(t)) - mg),\quad t \in [0,T]$$
$$x(0)=0$$,  
$$y(0)=1000, y(T)=900$$,  
$$v_x(0)=13.23, v_x(T)=13.23,$$,  
$$v_y(0)=-1.288, v_y(T)=-1.288$$,  
$$T \geq 0$$.  
where:  

$$n(t) = arctan(\frac{v_y(t)-u_a(x(t))}{v_x(t)}),$$
$$v_r(t) = (\sqrt{v_x(t)^2+(v_y(t)-u_a(x(t))^2},$$
$$L(t) = \frac{1}{2}c_L(t)pSv_r(t)^2,$$
$$D(t) = \frac{1}{2}(c_0(t)+k(c_L(t))^2)pS(v_r(t))^2,$$

### Problem Reformulation on fixed time interval  $t \in [0,1]$ ###   
$$maximize \quad x(1)$$
Subject to:  

$$\frac{d}{dt}x(t) = Tv_x(t),\quad t \in [0,1]$$
$$\frac{d}{dt}v_x(t) = \frac{T}{m}(-Lsin(n(t)) - Dcos(n(t))),\quad t \in [0,1]$$
$$\frac{d}{dt}y(t) = Tv_y(t),\quad t \in [0,1]$$
$$\frac{d}{dt}v_y(t) = \frac{T}{m}(Lcos(n(t)) - Dsin(n(t)) - mg),\quad t \in [0,1]$$
$$x(0)=0$$,  
$$y(0)=1000, y(1)=900$$,  
$$v_x(0)=13.23, v_x(1)=13.23,$$,  
$$v_y(0)=-1.288, v_y(1)=-1.288$$,  
$$T \geq 0$$.  
where:  

$$n(t) = arctan(\frac{v_y(t)-u_a(x(t))}{v_x(t)}),$$
$$v_r(t) = (\sqrt{v_x(t)^2+(v_y(t)-u_a(x(t))^2},$$
$$L(t) = \frac{1}{2}c_L(t)pSv_r(t)^2,$$
$$D(t) = \frac{1}{2}(c_0(t)+k(c_L(t))^2)pS(v_r(t))^2,$$

### Mathematical Solution:  ###   
In a static version of the glider problem (no thermal) the optimal control is found mathematically by minimizing the ratio of drag to lift: $min \space D/L$. This minimum occurs at $c_L = \sqrt{\frac{c_0}{k}} = 0.69862$. Using the fact that there is no accelaration in a static solution we have that $-\frac{v_x}{v_y} = \frac{L}{D} = 10.274$ and we have that $v_r = \sqrt{\frac{2mg}{pS\sqrt{c_D^2 + c_L^2}}} = 13.2901$. From this we are able to compute that $v_x = 13.23 and v_y = -1.288$. Hence the boundary conditions on the velocities in the dynamic version of the problem are equal to the constant values for the velocities in the static version of the problem. From this we deduce that the solution to the dynamic problem is static. We will use these values for $v_x, v_y, and c_L$ as initial guesses in our dynamic problem. 

```python
#Example Problem 3 (Glider)
 
from pyomo.environ import *
from pyomo.dae import *
import time
start_time = time.time()

model = ConcreteModel()
 
#tau 
model.tau = ContinuousSet(bounds = (0,1))
 

#Constants
model.m = 100
model.S = 14
model.rho = 1.13
model.g = 9.81
model.clmax = 1.4
model.c0 = 0.034
model.k = 0.069662
 

#Unknowns
model.x = Var(model.tau)
model.vx = Var(model.tau, initialize = 13.23)
model.y = Var(model.tau)
model.vy = Var(model.tau, initialize = -1.288)
model.cl = Var(model.tau, bounds = (0, model.clmax), initialize=0.69)
model.T = Var(within = PositiveReals)
 
#Derivatives
model.dxdt = DerivativeVar(model.x, wrt = model.tau)
model.dvxdt = DerivativeVar(model.vx, wrt = model.tau)
model.dydt = DerivativeVar(model.y, wrt = model.tau)
model.dvydt = DerivativeVar(model.vy, wrt = model.tau)
 

#Useful Functions
def ua(x):
    um = 2.5
    R = 100
    return um*(exp(-(x/R-2.5)**2))*(1-(x/R-2.5)**2)
 
def eta(model, tau):
    return atan((model.vy[tau] - ua(model.x[tau]))/model.vx[tau])
 
def vr2(model, tau): #v_r squared
    return (model.vx[tau])**2 + (model.vy[tau] - ua(model.x[tau]))**2

def L(model, tau):
    return .5*model.cl[tau]*model.rho*model.S*vr2(model,tau)

def D(model, tau): #formula different from notes; possible error?
    return .5*(model.c0 + model.k*((model.cl[tau])**2))*model.rho*model.S*vr2(model,tau)
     
#objective
def obj_rule(model):
    return model.x[1]
model.obj = Objective(rule = obj_rule, sense = maximize)
 
#constraint b
def velox(model, tau):
    if tau == 0:
        return Constraint.Skip
    return model.dxdt[tau] == model.T * model.vx[tau]
model.consb = Constraint(model.tau, rule = velox)
 
#constraint c
def accelx(model, tau):
    if tau == 0:
        return Constraint.Skip
    return model.dvxdt[tau] == (model.T/model.m)*((-L(model,tau)*sin(eta(model,tau)))
                - (D(model,tau)*cos(eta(model,tau))))
model.consc = Constraint(model.tau, rule = accelx)
 
#constraint d
def veloy(model, tau):
    if tau == 0:
        return Constraint.Skip
    return model.dydt[tau] == model.T * model.vy[tau]
model.consd = Constraint(model.tau, rule = veloy)
 
#constraint e
def accely(model, tau):
    if tau == 0:
        return Constraint.Skip
    return model.dvydt[tau] == (model.T/model.m)*((L(model,tau)*cos(eta(model,tau)))
                - (D(model,tau)*sin(eta(model,tau))) - model.m*model.g)
model.conse = Constraint(model.tau, rule = accely)
 
 
 
#initial/final conditions
def condi(model):
    yield model.x[0] == 0
    yield model.y[0] == 1000
    yield model.vx[0] == 13.23
    yield model.vy[0] == -1.288
    yield model.y[1] == 900
    yield model.vx[1] == 13.23
    yield model.vy[1] == -1.288
    yield ConstraintList.End
model.cond = ConstraintList(rule = condi)

#Finite Difference Discretizing
#discretizer = TransformationFactory('dae.finite_difference')
#discretizer.apply_to(model, nfe = 20, wrt = model.tau, scheme = 'BACKWARD')
 
#Collocation Discretizing
discretizer = TransformationFactory('dae.collocation')
discretizer.apply_to(model, nfe=10, ncp=5, scheme='LAGRANGE-RADAU')
 

solver = SolverFactory('ipopt')
solver.options['max_iter']= 500
 
results = solver.solve(model)
 
print("Optimal Horizontal Distance is %4.2f meters" %value(model.x[1]))
print("Optimal Time is %2.4f seconds" %value(model.T))
print("--- %s seconds ---" % (time.time() - start_time))
 
t = []
x = []
vx = []
y = []
vy = []
cl = []
 
for j in sorted(model.tau):
    tv = value(model.T) * j
    t.append(tv)
    x.append(value(model.x[j]))
    vx.append(value(model.vx[j]))
    y.append(value(model.y[j]))
    vy.append(value(model.vy[j]))
    cl.append(value(model.cl[j]))
     
import matplotlib.pyplot as plt
plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('Horizontal Position')
plt.title('x vs t')
plt.show()
 
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('Vertical Position')
plt.title('y vs t')
plt.show()
 
plt.plot(t,vx)
plt.xlabel('time')
plt.ylabel('Horizontal Velocity')
plt.title('vx vs t')
plt.show()
 
plt.plot(t,vy)
plt.xlabel('time')
plt.ylabel('Vertial Velocity')
plt.title('vy vs t')
plt.show()
 
plt.plot(t,cl)
plt.xlabel('time')
plt.ylabel('Lift Coefficient')
plt.title('cL vs t')
plt.show()
 
plt.plot(x,y)
plt.xlabel('Horizontal Position')
plt.ylabel('Vertical Position')
plt.title('y vs x')
plt.show()
```
