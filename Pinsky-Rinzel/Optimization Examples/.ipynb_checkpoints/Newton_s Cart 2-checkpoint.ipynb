# Newton's Cart Time Minimization #

Implementation of Newton Cart second example, where the goal is to push the cart 100m in as small as possible with the applied force restricted to be between $u_{min}$ and $u_{max}$

Problem formulation:  
$$minimize\quad T$$
Subject to  
$$\frac{d}{dt}y_1(t) = y_2(t), t \in [0,T]$$ 
$$\frac{d}{dt}y_2(t) = u(t) \div M, t \in [0,T]$$  
$$y_1(0) = 0, y_2(0) = 0, y_1(T) = 100,$$  
$$u(t) \in [u_{min},u_{max}], t \in [0,T]$$  
$$T \geq 0$$
         

Problem Reformulation on fixed time interval $t \in [0,1]$.
$$minimize\quad T$$
Subject to  
$$\frac{d}{dt}\tilde{y}_1(t) = T\tilde{y}_2(t), t \in [0,1]$$ 
$$\frac{d}{dt}\tilde{y}_2(t) = \frac{T}{M}\tilde{u}(t), t \in [0,1]$$  
$$\tilde{y}_1(0) = 0, \tilde{y}_2(0) = 0, \tilde{y}_1(1) = 100,$$  
$$\tilde{u}(t) \in [u_{min},u_{max}], t \in [0,1]$$  
$$T \geq 0$$  
where  
$$\tilde{u}(t) = u(Tt)$$
$$\tilde{y_1}(t) = y_1(Tt)$$
$$\tilde{y_2}(t) = y_2(Tt)$$  
and hence
$$\frac{d}{dt}\tilde{y_1}(t) = T\frac{d}{dt}y_1(Tt)$$
$$\frac{d}{dt}\tilde{y_2}(t) = T\frac{d}{dt}y_2(Tt)$$

```python
from pyomo.environ import *
from pyomo.dae import *
import matplotlib.pyplot as plt
infinity = float('inf')

model = ConcreteModel()

model.t = ContinuousSet(bounds=(0,1))
model.mass = Param(initialize=1.0)
model.Umin = Param(initialize=0.0)
model.Umax = Param(initialize=10.0)

model.tt = Var(within=PositiveReals)

model.x = Var(model.t, within=NonNegativeReals)
model.v = Var(model.t, within=NonNegativeReals)
model.u = Var(model.t, within=NonNegativeReals, bounds=(model.Umin,model.Umax))

model.dxdt = DerivativeVar(model.x, wrt=model.t, within=NonNegativeReals)
model.dvdt = DerivativeVar(model.v, wrt=model.t, within=NonNegativeReals)

model.x[0].fix(0.0)
model.x[1].fix(100.0)
model.v[0].fix(0.0)

def _ode1(m, t):
    if t == 100:
        return Constraint.Skip
    return m.dxdt[t] == m.tt * m.v[t]
model.ode1 = Constraint(model.t, rule=_ode1)

def _ode2(m, t):
    if t == 100:
        return Constraint.Skip
    return m.dvdt[t] == m.tt * m.u[t] / m.mass
model.ode2 = Constraint(model.t, rule=_ode2)

discretizer = TransformationFactory('dae.collocation')
discretizer.apply_to(model,nfe=20,ncp=6,scheme='LAGRANGE-RADAU')

def _obj1(m):
    return m.tt

model.obj1 = Objective(rule=_obj1)

solver = SolverFactory('ipopt')
results = solver.solve(model)

xf = []
vf = []
uf = []
tf = []

for i in sorted(model.t):
    tf.append(i)
    xf.append(value(model.x[i]))
    vf.append(value(model.v[i]))
    uf.append(value(model.u[i]))

plt.figure(1)
plt.plot(tf,xf)
plt.title('Position Results')
plt.xlabel('Time')
plt.ylabel('Position')

plt.figure(2)
plt.plot(tf,vf)
plt.title('Velocity Results')
plt.xlabel('Time')
plt.ylabel('Velocity')

plt.figure(3)
plt.plot(tf,uf)
plt.title('Applied Force Results')
plt.xlabel('Time')
plt.ylabel('Applied Force')

print(value(model.tt))
#model.solutions.store_to(results)
#results.write()
```
